```python
import pandas as pd
```

# loading data

```python
top100 = pd.read_csv("docleaderboard-top100.tsv", sep=' ', names=["qid", "Q0", "docid", "rank", "score", "link"])
```

```python
top100
```

```python
queries = pd.read_csv("docleaderboard-queries.tsv", sep='\t',names=["qid","query"])
```

```python
queries
```

```python
docs = pd.read_csv("msmarco-docs.tsv", sep='\t', names=["docid", "url", "title", "body"])
```

```python
docs
```

## merging top100 and queries

```python
top100_with_queries = pd.merge(top100, queries, on="qid", how="left")
```

```python
top100_with_queries
```

## merging top100_with_queries and docs

```python
data = pd.merge(top100_with_queries, docs, on="docid", how="left")
```

```python
data
```

## new data

```python
new_data_ = data[["qid","docid","query","title","body"]]
```

```python
new_data_
```

```python
new_data_.to_csv("../input_data/test_query.tsv",sep='\t',index=False)
```
